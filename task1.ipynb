```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("task 1.csv")
df
```

```python
# Bar chart to visualize age group distribution
plt.figure(figsize=(8, 6))
plt.bar(df["Age Group"], df["Population (Millions)"], color=["gold", "deepskyblue", "hotpink"])
plt.title("India's Population Distribution by Age in 2022")
plt.xlabel("Age Group")
plt.ylabel("Population (Millions)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
